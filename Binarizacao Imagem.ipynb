```python
#Projeto para redução de dimensionalidade em imagens.
# Seguindo o exemplo do algoritmo de binarização, realize a implementação em Python para transformar uma imagem colorida para
# - níveis de cinza (0 a 255) e para binarizada (0 e 255), preto e branco.
# Importando as bibliotecas
import cv2
import numpy as np
import urllib.request
from google.colab.patches import cv2_imshow
```

```python
# URL da imagem
url = 'https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg'

# Baixa e decodifica a imagem
resp = urllib.request.urlopen(url)
image_array = np.asarray(bytearray(resp.read()), dtype=np.uint8)
img = cv2.imdecode(image_array, cv2.IMREAD_COLOR)

# Verifica se carregou corretamente
if img is None:
    print("Erro ao carregar a imagem.")
```

```python
# 2. Reduz a imagem (opcional, se quiser diminuir a dimensão)
escala = 50  # porcentagem
if img is not None:
    largura = int(img.shape[1] * escala / 100)
    altura = int(img.shape[0] * escala / 100)
    dim = (largura, altura)
    img_redimensionada = cv2.resize(img, dim, interpolation=cv2.INTER_AREA)
else:
    print("Image not loaded, skipping resizing.")
    img_redimensionada = None # Set img_redimensionada to None if img is not loaded
```

```python
# 3. Converte para escala de cinza (0 a 255)
img_cinza = cv2.cvtColor(img_redimensionada, cv2.COLOR_BGR2GRAY)
```

```python
# 4. Aplica binarização (0 ou 255)
_, img_bin = cv2.threshold(img_cinza, 127, 255, cv2.THRESH_BINARY)
```

```python
# 5. Exibe os resultados
print("Imagem original redimensionada:")
cv2_imshow(img_redimensionada)

print("Imagem em escala de cinza:")
cv2_imshow(img_cinza)

print("Imagem binarizada:")
cv2_imshow(img_bin)
```
